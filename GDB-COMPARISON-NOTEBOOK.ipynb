```python
#TASK - REVIEW FIELDS IN LAYERS IN EACH DATABASE TO DETERMINE WHETHER FIELD MAPPING IS REQUIRED
#ThIS CELL REVIEWS EACH DATABASE - THE OLD AND THE NEW TEMPLATE DATABASE - AND LISTS OUT FIELDS INTO 
    

import arcpy
import pandas as pd

# Paths to the two geodatabases
gdb1_path = r'F:\ROW\Python_GDB_Conversion\SAAN_LANDS_ProjectData.gdb\SAAN_LANDS_ProjectData'
gdb2_path = r'F:\ROW\Python_GDB_Conversion\SAAN.gdb\SAAN_Features'

# Function to list fields for a given feature class"C:\\Users\\hdean\\Desktop\\GDB_Conversion_2024\\Appending_Old_to_New_Layers_V3_06212024.ipynb"
def list_fields(layer):
    fields = arcpy.ListFields(layer)
    return [field.name for field in fields]

try:
    # Set workspace to the first geodatabase
    arcpy.env.workspace = gdb1_path
    
    # Create an empty list to store field data
    data_gdb1 = []
    # List all feature classes in the first geodatabase
    for fc in arcpy.ListFeatureClasses():
        # List fields for the current feature class
        fields = list_fields(fc)
        # Append feature class and its fields to the data list
        data_gdb1.append({'Feature Class': fc, 'Fields': fields})
    
    # Set workspace to the second geodatabase
    arcpy.env.workspace = gdb2_path
    
    # Create an empty list to store field data
    data_gdb2 = []
    # List all feature classes in the second geodatabase
    for fc in arcpy.ListFeatureClasses():
        # List fields for the current feature class
        fields = list_fields(fc)
        # Append feature class and its fields to the data list
        data_gdb2.append({'Feature Class': fc, 'Fields': fields})
    
    # Convert lists to Pandas DataFrames
    df_gdb1 = pd.DataFrame(data_gdb1)
    df_gdb2 = pd.DataFrame(data_gdb2)
    
    # Display the DataFrames
    print("Fields in Geodatabase 1:")
    print(df_gdb1)
    
    print("\nFields in Geodatabase 2:")
    print(df_gdb2)

    print("\nField listing complete.")

except arcpy.ExecuteError:
    print(arcpy.GetMessages(2))

except Exception as ex:
    print(ex)
```

```python
# TASK - PRINT PD's to EXCEL FOR CROSSWALK EXERCISE
# Paths to the two geodatabases
import arcpy
import pandas as pd

gdb1_path = r'F:\ROW\Python_GDB_Conversion\SAAN_LANDS_ProjectData.gdb\SAAN_LANDS_ProjectData'
gdb2_path = r'F:\ROW\Python_GDB_Conversion\SAAN.gdb\SAAN_Features'

# Function to list fields for a given feature class
def list_fields(layer):
    fields = arcpy.ListFields(layer)
    return [field.name for field in fields]

try:
    # Create an empty DataFrame to store all fields
    df_all_fields = pd.DataFrame(columns=['Feature Class', 'Field'])
    
    # Set workspace to the first geodatabase
    arcpy.env.workspace = gdb1_path
    
    # List all feature classes in the first geodatabase
    for fc in arcpy.ListFeatureClasses():
        # Create DataFrame for current feature class in gdb1
        df_gdb1 = pd.DataFrame({'Field': list_fields(fc)})
        # Add feature class name as a new column
        df_gdb1['Feature Class'] = fc
        # Append DataFrame to the master DataFrame
        df_all_fields = pd.concat([df_all_fields, df_gdb1], ignore_index=True)
    
    # Set workspace to the second geodatabase
    arcpy.env.workspace = gdb2_path
    
    # List all feature classes in the second geodatabase
    for fc in arcpy.ListFeatureClasses():
        # Create DataFrame for current feature class in gdb2
        df_gdb2 = pd.DataFrame({'Field': list_fields(fc)})
        # Add feature class name as a new column
        df_gdb2['Feature Class'] = fc
        # Append DataFrame to the master DataFrame
        df_all_fields = pd.concat([df_all_fields, df_gdb2], ignore_index=True)
    
    # Write DataFrame to CSV file
    df_all_fields.to_csv(r"F:\ROW\Python_GDB_Conversion\gdbs_all_fields.csv", index=False)
    

except arcpy.ExecuteError:
    print(arcpy.GetMessages(2))

except Exception as ex:
    print(ex)
```

```python
#TASK - CREATE A CELL THAT TAKES AN EXCEL LIST OF OLD AND NEW LAYERS, ACCESS THE FIELDS IN EACH LAYER
    # AND PRINTS OUT A COMPARISON OF THE FIELDS AS A LIST OF UNCOMMON FIELDS
    
import arcpy
import pandas as pd

# Paths to the two geodatabases
gdb1_path = r'F:\ROW\Python_GDB_Conversion\SAAN_LANDS_ProjectData.gdb\SAAN_LANDS_ProjectData'
gdb2_path = r'F:\ROW\Python_GDB_Conversion\SAAN.gdb\SAAN_Features'
csv_file_path = r'F:\ROW\Python_GDB_Conversion\CSV_LAYER_COMPARE_LIST.csv'

# Function to list fields for a given feature class
def list_fields(layer):
    fields = arcpy.ListFields(layer)
    return [field.name for field in fields]

try:
    # Read the Excel file into a DataFrame
    df = pd.read_csv(csv_file_path)
    
    # Iterate through each row in the DataFrame
    for index, row in df.iterrows():
        old_layer = row['Layer_old']
        new_layer = row['Layer_new']
        
        # Set workspace to the first geodatabase
        arcpy.env.workspace = gdb1_path
        old_layer_path = f'{gdb1_path}\\{old_layer}'
        
        # Set workspace to the second geodatabase
        arcpy.env.workspace = gdb2_path
        new_layer_path = f'{gdb2_path}\\{new_layer}'
        
        # Check if both layers exist
        if arcpy.Exists(old_layer_path) and arcpy.Exists(new_layer_path):
            # Get fields for old and new layers
            old_fields = set(list_fields(old_layer_path))
            new_fields = set(list_fields(new_layer_path))
            
            # Compare fields
            common_fields = old_fields.intersection(new_fields)
            unique_fields_old = old_fields - new_fields
            unique_fields_new = new_fields - old_fields
            
            # Print comparison results
            print(f"Comparison for '{old_layer}' and '{new_layer}':")
            print(f"Common fields: {common_fields}")
            print(f"Fields unique to '{old_layer}': {unique_fields_old}")
            print(f"Fields unique to '{new_layer}': {unique_fields_new}")
            print("\n")
        else:
            print(f"One or both layers '{old_layer}' and '{new_layer}' do not exist.")

    print("Field comparison completed successfully.")

except Exception as ex:
    print(ex)
```

```python
# TASK - REVIEW DOMAINS IN TWO GDBs IN ORDER TO COMPARE

import arcpy
import pandas as pd

# Set paths to the geodatabases
gdb1_path = r'F:\ROW\Python_GDB_Conversion\SAAN_LANDS_ProjectData.gdb'
gdb2_path = r'F:\ROW\Python_GDB_Conversion\SAAN.gdb'

# Function to extract domain information and load it into a DataFrame
def extract_domain_info(gdb_path):
    arcpy.env.workspace = gdb_path
    domains = arcpy.da.ListDomains()

    domain_info = []
    for domain in domains:
        domain_name = domain.name
        domain_type = domain.domainType
        coded_values = domain.codedValues if domain_type == 'CodedValue' else None
        range_info = (domain.minValue, domain.maxValue) if domain_type == 'Range' else None
        domain_info.append({'Name': domain_name, 'Type': domain_type, 'CodedValues': coded_values, 'RangeInfo': range_info})

    return pd.DataFrame(domain_info)

# Extract domain information from GDB1 and GDB2
df_gdb1_domains = extract_domain_info(gdb1_path)
df_gdb2_domains = extract_domain_info(gdb2_path)

# Compare domain information between the two geodatabases
domain_comparison = pd.merge(df_gdb1_domains, df_gdb2_domains, on='Name', how='outer', suffixes=('_GDB1', '_GDB2'))

# Display the comparison
print(domain_comparison)


# Function to extract and print domain information
def print_domain_info(gdb_path):
    arcpy.env.workspace = gdb_path
    domains = arcpy.da.ListDomains()

    for domain in domains:
        print(f"Domain Name: {domain.name}")
        print(f"  Type: {domain.domainType}")
        if domain.domainType == 'CodedValue':
            print("  Coded Values:")
            for code, value in domain.codedValues.items():
                print(f"    {code}: {value}")
        elif domain.domainType == 'Range':
            print(f"  Range: {domain.minValue} - {domain.maxValue}")
        print()

# Print domain information for GDB1 and GDB2
print("Domain Information for GDB1:")
print_domain_info(gdb1_path)

print("Domain Information for GDB2:")
print_domain_info(gdb2_path)
```

```python
# TASK - PLACE DOMAINS AND LISTS INTO CSV FILES FOR REVIEW

# Set paths to the geodatabases and CSV files
gdb1_path = r'F:\ROW\Python_GDB_Conversion\SAAN_LANDS_ProjectData.gdb'
gdb2_path = r'F:\ROW\Python_GDB_Conversion\SAAN.gdb'
csv1_path = r"F:\ROW\Python_GDB_Conversion\path_to_csv1.csv"
csv2_path = r"F:\ROW\Python_GDB_Conversion\path_to_csv2.csv"

# Function to extract domain information and save it to CSV
def extract_and_save_domains(gdb_path, csv_path):
    arcpy.env.workspace = gdb_path
    domains = arcpy.da.ListDomains()

    domain_info = []
    for domain in domains:
        domain_name = domain.name
        domain_type = domain.domainType
        coded_values = domain.codedValues if domain_type == 'CodedValue' else None
        range_info = (domain.minValue, domain.maxValue) if domain_type == 'Range' else None
        domain_info.append({'Name': domain_name, 'Type': domain_type, 'CodedValues': coded_values, 'RangeInfo': range_info})

    df = pd.DataFrame(domain_info)
    df.to_csv(csv_path, index=False)

# Extract and save domain information from GDB1 and GDB2
extract_and_save_domains(gdb1_path, csv1_path)
extract_and_save_domains(gdb2_path, csv2_path)

print("Domain information saved to CSV files.")
```
